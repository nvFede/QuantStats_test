```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
import quantstats as qs
```

```python
data = yf.download('AAPL', auto_adjust=True)
```

```python
data['pct_change'] = data.Close.pct_change() * 100
```

```python
data['volatilidad_anualizada'] = data['pct_change'].rolling(252).std() * 252**0.5
```

```python
data.tail()
```

```python
data['pct_change'].plot()
```

```python
data[ 'volatilidad_anualizada'].plot()
```
